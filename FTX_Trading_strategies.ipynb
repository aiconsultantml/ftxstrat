```python
#Only one cell for easy execution

!pip install --quiet yfinance ta  pysftp prophet


path = '/content/drive/MyDrive/quantalgo/crypto/'



#------------------------------------------------------------------------------------
#Last Buy and Sell signals

from requests import  Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import pandas as pd
import datetime

import json
from requests import  Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import pandas as pd
import numpy as np
import yfinance as yf
import ta
from prophet import Prophet
import datetime
import time
import pysftp
import math


import warnings
warnings.filterwarnings("ignore")

#path = "data/"


#*****************************
#** This function is for the cryptos that are less than $1
#** provides a multiplier and zero padded value
#** 0.005 will be 500.0, with a multiplier 100000
#*****************************

def get_zero_padding(val):

  val_log = math.log10(val)
  print("val_log: ", val_log)

  val_pad, r=(divmod(val_log, 1)) 

  val_pad = math.floor(val_log // 1)
  print("val_pad: ", val_pad)

  strn = '{:<0' + str(abs(val_pad)+1) + 'd}'

  print("strn: ", strn)
  #strn = '{:<04d}'
  print("strn: ", strn.format(1))

  print("zero_padded_val ",int(strn.format(1)) * val)

  
  zero_padding_mult = int(strn.format(1)) * 100
  zero_padded_val = zero_padding_mult * val
  return zero_padded_val, zero_padding_mult

#*****************************
#** Calculator for trade_ratio
#*****************************

def calc_trade_ratio(signal_forecast_df):



  #**internal function to calculate 15 days, 30 days trade ratio
  def sub_calc_trade_ratio(a_df):

    sell_cnt = a_df[a_df.final_alert=="SELL"].alert.count()
    buy_cnt = a_df[a_df.final_alert=="BUY"].alert.count()

    #print("before sell count", sell_cnt)
    #print("before buy count", buy_cnt)


    if sell_cnt > buy_cnt:
      a_df.drop(a_df[a_df.Close == a_df[a_df.final_alert=="SELL"].Close.min()].index, inplace = True)

    if sell_cnt < buy_cnt:
      a_df.drop(a_df[a_df.Close == a_df[a_df.final_alert=="BUY"].Close.max()].index, inplace = True)

    sell_cnt = a_df[a_df.final_alert=="SELL"].alert.count()
    buy_cnt = a_df[a_df.final_alert=="BUY"].alert.count()

    #print("after sell count", sell_cnt)
    #print("after buy count", buy_cnt)


    temp_df=a_df
    temp_df['trade'] = np.where(temp_df.final_alert=="BUY", temp_df.Close, np.where(temp_df.final_alert=="SELL",temp_df.Close*-1,0) )

    trade_ratio = abs(temp_df.trade.sum())
    #print(trade_ratio)
    return trade_ratio


  #**main statements for this function  
  temp_df = signal_forecast_df
  temp_df['alert'] = np.where(
                                ((temp_df.final_signal=="BUY") & (temp_df.Close < temp_df.yhat)), temp_df.final_signal, 
                                  np.where(  ((temp_df.final_signal=="SELL") & (temp_df.Close > temp_df.yhat)), temp_df.final_signal, "-")
                    )
  temp_df = temp_df[ (temp_df.alert == "BUY") | (temp_df.alert == "SELL") ] 
  temp_df['alert_prev'] = temp_df.alert.shift(1)
  temp_df['final_alert'] = np.where((temp_df.alert != temp_df.alert_prev), temp_df.alert , "-" )

  #this dataframe is to hold the last 5 alerts for the cryptos
  last_5_alerts_df = temp_df[-5:]

  #print(temp_df.groupby(['final_alert'], sort=False)['Datetime'].count())
  a_df = temp_df[1:]
  start_date_7 = datetime.date.today() - datetime.timedelta(days=8)   #For last 7 days 
  start_date_7 = start_date_7.strftime("%Y-%m-%d %H:%M:%S")

  trade_ratio_7 = sub_calc_trade_ratio(a_df[a_df.Datetime >= start_date_7])
  print("a_df_7 ", len(a_df[a_df.Datetime >= start_date_7])) 


  start_date_30 = datetime.date.today() - datetime.timedelta(days=31)   #For last 30 days 
  start_date_30 = start_date_30.strftime("%Y-%m-%d %H:%M:%S")

  trade_ratio_30 = sub_calc_trade_ratio(a_df[a_df.Datetime >= start_date_30])
  print("a_df_30 ", len(a_df[a_df.Datetime >= start_date_30])) 
  

  return(trade_ratio_30, trade_ratio_7, last_5_alerts_df)


#*****************************
#Fetch All Data
#*****************************

def fetchCryptoDataCoingecko(SymbolNameString):
    #### PROD URL
    #url = 'https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&symbol={}&order=market_cap_desc&per_page=100&page=1&sparkline=true'.format(SymbolNameString.lower())
    #print(url)
    #url = 'https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc&sparkline=true'
    url = 'https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc&sparkline=false&per_page=110&ids='+SymbolNameString

    session = Session()
      
    try:
        response = session.get(url)
        data = response.json() 
        #print(data) 
        df = pd.DataFrame(data)
        
        for info in data:            
            
            cryptoVolume = float(info['total_volume'])
            cryptoMarketCap = float(info['market_cap'])
            #print(cryptoVolume, cryptoMarketCap)
            
    except (ConnectionError, Timeout, TooManyRedirects) as e:
        print(e)
    
    return(df)

#*****************************************************************************
#-------------------- NEW ML MODEL ----------------------------------------
#*****************************************************************************

def get_data_signal(start_dt, end_dt, ticker, interval):
    
    df = yf.download(ticker, start=start_dt, end=end_dt, interval=interval)


    df.reset_index(level=0, inplace=True)
    print(ticker, " yf length: ", len(df))
    #df.head()
    df['ta_rsi'] = ta.momentum.rsi(df.Close, window=14)
    df['ta_rsi1'] = df.ta_rsi.shift(1)

    df['ta_macd'] = ta.trend.MACD(df.Close, window_slow = 26, window_fast = 12, window_sign = 9, fillna = False).macd()
    df['ta_macd_signal'] = ta.trend.MACD(df.Close, window_slow = 26, window_fast = 12, window_sign = 9, fillna = False).macd_signal()
    df['ta_macd_diff'] = ta.trend.MACD(df.Close, window_slow = 26, window_fast = 12, window_sign = 9, fillna = False).macd_diff()
    #decide_action
    df['ta_macd_diff'] = np.where(df.ta_macd > df.ta_macd_signal, 1, -1) 
    df['macd_cal_5'] = np.where( (df.ta_macd_diff>=1), 1, 0 )  
    df['macd_cal_n5'] = np.where((df.ta_macd_diff<=-1), 1, 0 )

    df['ta_macd_diff1'] = df.ta_macd_diff.shift(1)
    df['macd_reversal'] = np.where((df.ta_macd_diff + df.ta_macd_diff1 == 0), 1, 0 )


    df['rsi_calc'] = np.where(
                          ((df.ta_rsi < 81) & (df.ta_rsi > 60)) | 
                            ((df.ta_rsi < 41) & (df.ta_rsi > 20)), 
                              1, 0 )

    df['rsi_prev'] = np.where(df.rsi_calc==1,
                              np.where(df.ta_rsi1 < df.ta_rsi , 1, -1 ), 
                              0)

    df['signal'] = np.where(
                              ((df.rsi_calc==1) & (df.macd_cal_5 ==1) & ((df.rsi_prev ==1) | (df.macd_reversal ==1)) ), "BUY", 
                              np.where(  ( (df.rsi_calc==1) &  (df.macd_cal_n5 == 1) & ((df.rsi_prev ==-1) | (df.macd_reversal ==1)) ), "SELL", "-")
                            )

    #df = df.rename(columns={"index" : "dt"})
    df.Datetime = df.Datetime.astype(str).str[:-6]

    df['signal_prev'] = df.signal.shift(1)
    df['final_signal'] = np.where((df.signal != "-") & (df.signal_prev=="-"), df.signal , "-" )

    df = df.drop(columns = ['ta_rsi1'])

    return df

def get_signal_forecast(start_dt, end_dt, ticker):
      
    df_30 = get_data_signal(start_dt, end_dt, ticker, "30m")
    df_15 = get_data_signal(start_dt, end_dt, ticker, "15m")

    #This statement join the df_30 and df_15 on Datetime, it gives _x and _y
    df = pd.merge(df_30, df_15, on='Datetime', how='left')

    #This statement compares the Final_Signal from  df_30 and df_15 
    df.signal_x = df.final_signal_x
    df.signal_prev_x = df.final_signal_y #to avoid increase of unnecessary cols, replace signal_prev with df_15.Final_Signal

    df['final_signal_x'] = np.where(df.final_signal_x == "-", df.final_signal_y, df.final_signal_x)
    
    #this statement rips _x from the first set of rows
    df.columns = df.columns.str.rstrip('_x')

    #this statements drops all the colummns for the df_15
    df = df[df.columns.drop(list(df.filter(regex='_y')))]
    df['final_signal'] = np.where((df.signal != "-") & (df.signal_prev=="-"), df.signal , "-" )
    #print(df)


    #-----------------------------
    #calculate for varying dollars - start
    #-----------------------------
    print("last record close ", df.Close.max())

    adjust_close_lt1_flag = "N"
    adjust_close_1to100_flag = "N"
    zero_padded_val, zero_padding_mult = 0, 0

    if df.Close.max() < 1:
        zero_padded_val, zero_padding_mult = get_zero_padding(df.Close.max())
        df.Close = df.Close*zero_padding_mult
        adjust_close_lt1_flag = "Y"
        print("here -1a \n \n ")

    if 1 <= df.Close.max() <= 99:
        df.Close = df.Close*100
        adjust_close_1to100_flag = "Y"
        print("here -1b \n \n ")


    #-----------------------------------------------------------------------------------
    #Automated PROPHET
    #-----------------------------------------------------------------------------------

    cryto_df = df.iloc[:-1,[0,4]]  
    cryto_df = cryto_df.rename(columns={'Datetime':'ds','Close':'y'})
    
    m = Prophet(interval_width=0.95)
    m.fit(cryto_df)

    future = m.make_future_dataframe(periods=4, freq='30min')

    forecast = m.predict(future)
    forecast = forecast.round(0)

    #merge forecase and signal df
    forecast1 = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
    forecast1 = forecast1.rename(columns={'ds': 'Datetime'})
    


    df.Datetime = df.Datetime.astype('datetime64[ns]')
  
    #print("forecast: \n",forecast1)

    signal_forecast_df = pd.merge( forecast1, df, on="Datetime", how ="left")
    signal_forecast_df = signal_forecast_df.assign(ticker=ticker)
    
    
    #calculate trade ratio for 30 and 7 days
    trade_ratio_30, trade_ratio_7, last_5_alerts_df = calc_trade_ratio(signal_forecast_df)

    signal_forecast_df = signal_forecast_df.assign(trade_ratio_30=trade_ratio_30)
    signal_forecast_df = signal_forecast_df.assign(trade_ratio_7=trade_ratio_7)

    #print("signal_forecast_df: \n",signal_forecast_df)

    #-----------------------------
    #calculate for varying dollars - end
    #-----------------------------
    
    if adjust_close_lt1_flag == "Y":
      signal_forecast_df.Close = signal_forecast_df.Close / zero_padding_mult
      signal_forecast_df.yhat = signal_forecast_df.yhat / zero_padding_mult
      signal_forecast_df.yhat_lower = signal_forecast_df.yhat_lower / zero_padding_mult
      signal_forecast_df.yhat_upper = signal_forecast_df.yhat_upper / zero_padding_mult
      signal_forecast_df.trade_ratio_30 = signal_forecast_df.trade_ratio_30 / zero_padding_mult
      signal_forecast_df.trade_ratio_7 = signal_forecast_df.trade_ratio_7 / zero_padding_mult

      
      #-------------------------------Last 5 Dataframe
      last_5_alerts_df.Close = last_5_alerts_df.Close / zero_padding_mult
      last_5_alerts_df.yhat = last_5_alerts_df.yhat / zero_padding_mult
      last_5_alerts_df.yhat_lower = last_5_alerts_df.yhat_lower / zero_padding_mult
      last_5_alerts_df.yhat_upper = last_5_alerts_df.yhat_upper / zero_padding_mult

        
                                      
      print("here -2a ")
    if adjust_close_1to100_flag == "Y":
        signal_forecast_df.Close = signal_forecast_df.Close / 100
        signal_forecast_df.yhat = signal_forecast_df.yhat / 100
        signal_forecast_df.yhat_lower = signal_forecast_df.yhat_lower / 100
        signal_forecast_df.yhat_upper = signal_forecast_df.yhat_upper / 100
        signal_forecast_df.trade_ratio_30 = signal_forecast_df.trade_ratio_30 / 100
        signal_forecast_df.trade_ratio_7 = signal_forecast_df.trade_ratio_7 / 100


      #-------------------------------Last 5 Dataframe
        last_5_alerts_df.Close = last_5_alerts_df.Close / 100
        last_5_alerts_df.yhat = last_5_alerts_df.yhat / 100
        last_5_alerts_df.yhat_lower = last_5_alerts_df.yhat_lower / 100
        last_5_alerts_df.yhat_upper = last_5_alerts_df.yhat_upper / 100
      
        print("here -2b ")

    print(ticker, " data length: ", len(signal_forecast_df))
    print(signal_forecast_df.tail(1))

    return (signal_forecast_df, last_5_alerts_df)

#*****************************************************************************
#-------------------- CALL PREDICTION MODEL --------------------------------
#*****************************************************************************
def predict_data(prediction_flag):
    #first logic to get top 100
    start_dt = "2022-10-31"
    end_dt = "2022-12-31" 

    crypto_symbols_df = pd.read_csv(path+"ftx_strategy.csv")
    symbols = crypto_symbols_df.id.to_string(index=False).replace("\n", ",").strip() 
    symbols = symbols.replace(" ", "").strip() 
    crypto_symbols_df = crypto_symbols_df.drop(columns=['symbol', 'name', 'current_price'])

    #symbol= "eth, btc, perp" 
    #call the coingecko function
    df = fetchCryptoDataCoingecko(symbols)

    df = pd.merge(crypto_symbols_df, df, how="right", on="id") 
    #df.to_csv(path+"dump_data.csv")
    #df['ticker'] = df.symbol.str.upper()
    #df['ticker'] = df.ticker.astype(str) + "-USD"
    
    print("PRINTING HERE \n", df.head())


    #stocks=['ETH-USD', 'BTC-USD', 'LTC-USD',  'DOGE-USD', 'BNB-USD', 'XLM-USD', 'ADA-USD',  'XRP-USD'] #'SHIB-USD', 'DOT-USD',
    #stocks=[ 'PERP-USD', 'BTC-USD', 'ETH-USD'] #'SHIB-USDT', 'DOT-USD',
    #print(type(stocks))

    #prediction_flag = "Y"
    i = 0
    if(prediction_flag == "Y"):
      all_df = pd.DataFrame()
      all_last_5_alerts_df = pd.DataFrame()

      for ticker in df['ticker']:
        print(i, " :################################ ")
        print("Processing: ", ticker)
        i = i+1
        try:
          signal_forecast_df, last_5_alerts_df = get_signal_forecast(start_dt, end_dt, str(ticker))
          all_df = all_df.append(signal_forecast_df)
          all_last_5_alerts_df = all_last_5_alerts_df.append(last_5_alerts_df)

          print("all_df intermediate len: ", len(all_df))
          print("all_last_5_alerts_df intermediate len: ", len(all_last_5_alerts_df))
        except:
          continue
        print("################################\n \n \n ")

      pred_datetime = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
      all_df = all_df.assign(pred_datetime=pred_datetime)    
      print("all_df len: ", len(all_df))

      crypto_symbols_df = pd.read_csv(path+"crypto_symbols.csv")
      all_last_5_alerts_df = pd.merge(crypto_symbols_df, all_last_5_alerts_df, how="right", on="ticker")
      print("all_last_5_df len: ", len(all_last_5_alerts_df))

      all_df.to_pickle(path+"all_df.pkl")
      all_last_5_alerts_df.to_pickle(path+"all_last_5_alerts_df.pkl")
      time.sleep(2)

    all_df = all_df.drop_duplicates(keep='first')
    all_df = pd.read_pickle(path+"all_df.pkl")
    all_df.to_csv(path+"all_df.csv")

    all_last_5_alerts_df = pd.read_pickle(path+"all_last_5_alerts_df.pkl")
    
    
    #print("all_df \n",all_df.head(20))
    temp_df = all_df
    temp_df = temp_df[(temp_df.final_signal == "-") | (temp_df.final_signal == "BUY") | (temp_df.final_signal == "SELL") ] 

    temp_df = temp_df.groupby(['ticker'], sort=False)['Datetime'].max()
    print("temp_df  \n", temp_df.head())
    print("df again-2  \n", df.head())
    
    crypto_master_df = pd.merge(df, temp_df, how="left", on="ticker")
    print("crypto_master len 1: ", len(crypto_master_df))


    print("HERE-1 at Crypto master df 1 \n", crypto_master_df)
    print("HERE-2 at all_df 1 \n", all_df)

    crypto_master_df = pd.merge(crypto_master_df, all_df, how="left", on=["ticker", "Datetime" ])
    print("crypto_master len 2: ", len(crypto_master_df))

    crypto_master_df['alert'] = np.where(
                                        ((crypto_master_df.final_signal=="BUY") & (crypto_master_df.Close < crypto_master_df.yhat)), crypto_master_df.final_signal, 
                                          np.where(  ((crypto_master_df.final_signal=="SELL") & (crypto_master_df.Close > crypto_master_df.yhat)), crypto_master_df.final_signal, "-")
                            )

    crypto_master_df = crypto_master_df.fillna(0)
    #crypto_master_df = crypto_master_df.drop_duplicates(keep='first')
    #print("HERE-2 at Crypto master df 2 \n", crypto_master_df)
    return crypto_master_df, all_last_5_alerts_df





#MAIN FUNCTION IS THIS
def refresh_data(flag):
  
  if flag =="base_data":
    df, all_last_5_alerts_df = predict_data("N")

  if flag =="predict":
     df, all_last_5_alerts_df = predict_data("Y")
      

  df.to_csv(path+"prediction.csv")
  print("Here ", df) 
  all_last_5_alerts_df.to_csv(path+"all_last_5_alerts_df.csv") 

  return("Data Refresh Success")
```
